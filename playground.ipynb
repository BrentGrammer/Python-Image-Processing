```python
# Golden Ratio formula with variable

import sympy as sym

x = sym.Symbol('x')
fx = (1 + sym.sqrt(x)) / 2

y_at_5 = fx.subs(x, 5)
display(y_at_5)

p1 = sym.plot(fx, markers=[{'args' : [5, y_at_5, 'r*'], 'ms' : 10, "label": "a"}], legend=True, label='f(x) = (1 + √x) / 2',show=False)
p1.extend(sym.plot(y_at_5, markers=[{'args': [5, y_at_5, 'go']}], show=False, label=f'Golden Ratio ≈ {y_at_5.evalf():.4f}'))

# p1[0].label = "a"
# p1[1].label = "b"

p1.legend = True

p1.show()
```

```python

sym.var("x")
p = sym.plot(x,-x, markers=[{'args' : [5, 5, 'r*'], 'ms' : 10, "label": "a"},
                      {'args' : [5, -5, 'r*'],'ms' : 10, "label": "b"}], show=False, legend=True)
p[0].label = "a"
p[1].label = "b"
p.show()
```
