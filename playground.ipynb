```python
# Golden Ratio formula with variable

import sympy as sym

x = sym.Symbol('x')
fx = (1 + sym.sqrt(x)) / 2

y_at_5 = fx.subs(x, 5)
display(y_at_5)

p1 = sym.plot(fx, label='f(x) = (1 + √x) / 2',show=False)
p1.extend(sym.plot(y_at_5, markers=[{'args': [5, y_at_5, 'go']}], show=False, label=f'Golden Ratio ≈ {y_at_5.evalf():.4f}'))
p1.legend = True

p1.show()
```
